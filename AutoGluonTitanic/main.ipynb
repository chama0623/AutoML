# AutoGluonでTitanicを実行してみる

AutoGluon : 簡単に使えるAutoMLツール. Amazonが開発しているがAWS環境は必要なく, Python環境で実行できる.


install
```
pip install autogluon
```

````python
from autogluon.tabular import TabularDataset, TabularPredictor

train_data = TabularDataset("./data/train.csv")
test_data = TabularDataset("./data/test.csv")
predictor = TabularPredictor(label='Survived').fit(train_data=train_data)
y_pred = predictor.predict(test_data)
````
